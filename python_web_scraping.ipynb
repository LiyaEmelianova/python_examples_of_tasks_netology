## Задание

Вам необходимо написать функцию, которая будет основана на поиске по сайту habr.com. Функция в качестве параметра должна принимать список запросов для поиска (например, ['python', 'анализ данных']) и на основе материалов, попавших в результаты поиска по каждому запросу, возвращать датафрейм вида:

<дата> - <заголовок> - <ссылка на материал> - <количество лайков>

В рамках задания предполагается работа только с одной (первой) страницей результатов поисковой выдачи для каждого запроса. Материалы в датафрейме не должны дублироваться, если они попадали в результаты поиска для нескольких запросов из списка.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def web_scraping(q, page):

    results = []

    for p in range(1, page +1):

        for query in q:

            ВASE_URL = 'https://habr.com'
            resp = requests.get(ВASE_URL + f'/ru/search/page{page}/',\
                                params = {'q': query, \
                                          'target_type': 'posts',\
                                          'order': 'relevance'})
            soup = BeautifulSoup(resp.text)
            # print(soup.prettify())

            for article_tag in soup.find_all('article'):
                datetime = article_tag.find('time').get('datetime')
                h2_tag = article_tag.find('h2')
                title = h2_tag.get_text()
                try:
                    link = h2_tag.find('a').get('href')
                except:
                  header_tag = article_tag.find('header')
                  link = header_tag.find('a').get('href')
                rating = article_tag.find('span',\
                                          class_ = 'tm-votes-meter__value'\
                                          ).get_text()

                results.append([datetime, title, link, rating])

        habr_search_df = pd.DataFrame(results, \
                                      columns=['Date', 'Title',\
                                               'Link', 'Rating'])

        habr_search_df.Date = pd.to_datetime(habr_search_df.Date).dt.date
        habr_search_df.Link = ВASE_URL + habr_search_df.Link
        habr_search_df.drop_duplicates(inplace = True)

        return habr_search_df


keywords = ['python', 'анализ данных']
page = 1

web_scraping(keywords, page)
```
